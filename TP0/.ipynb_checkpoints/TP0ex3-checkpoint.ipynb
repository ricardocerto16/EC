<b> Exercício 3 - Corpos finitos primos </b>

Para realizarmos este exercício tivemos que adquirir o conceito do que são corpos finitos. Depois de adquirirmos este conceito, dividimos o exercício em três fases distintas. Numa primeira fase procedemos à criação dos corpos finitos primos, na segunda fase para cada um dos corpos finitos criamos o respetivo plot da função $x -> x^2$  e por fim na terceira e última fase determinamos um elemento primitivo para cada um dos corpos que verificasse uma determinada proposição.

**3 a)** <br/>
É pedido, inicialmente, que sejam criados quatro corpos finitos primos, sendo eles p=31, 127, 8191, 131071, recorrendo à função do SageMath GF(p), em que p é um número primo.

```python
p1 = 31
corpo1 = GF(p1)
print(corpo1)

p2 = 127
corpo2 = GF(p2)
print(corpo2)

p3 = 8191
corpo3 = GF(p3)
print(corpo3)

p4 = 131071
corpo4 = GF(p4)
print(corpo4)
```

**3 b)** <br/>
De seguida, foi criado um gráfico para cada corpo criado, através da ferramenta do SageMath plot. De modo a que o gráfico aparecesse, utilizou-se a função plot.show(). 

```python
import matplotlib.pyplot as plt
grafico=plt.plot([x**2 for x in corpo1])
plt.show(grafico)
```

```python
grafico=plt.plot([x**2 for x in corpo2])
plt.show(grafico)
```

```python
grafico=plt.plot([x**2 for x in corpo3])
plt.show(grafico)
```

```python
grafico=plt.plot([x**2 for x in corpo4])
plt.show(grafico)
```

Depois de analisar os resultados obtidos e respetivos gráficos, podemos verificar que as suas representações encontram-se corretas. Por exemplo, para o caso do corpo finito com o número primo 31, podemos averiguar que para x=5, então 5x5=25. Para 6x6=36. No entanto 36, como excede o número primo 31, o valor será a diferença dos dois números (36-31)=5 como se pode confirmar pelo gráfico.

**3 c)** <br/>

No último ponto, determina-se um elemento primitivo para cada um dos corpos finitos primos, depois verifica-se a proposição exposta no enunciado para cada um desses elementos primitivos, da qual se apresenta sempre verdadeira.

```python
e1 = corpo1.primitive_element()
print(e1)
n = mod(0,p1-1)
e1^n == 1
```

```python
e2 = corpo2.primitive_element()
print(e2)
n = mod(0,p2-1)
e2^n == 1
```

```python
e3 = corpo3.primitive_element()
print(e3)
n = mod(0,p3-1)
e3^n == 1
```

```python
e4 = corpo4.primitive_element()
print(e4)
n = mod(0,p4-1)
e4^n == 1
```
